```python
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame containing the dataset
plt.hist(df['stars'], bins=5, color='skyblue', edgecolor='black')
plt.title('Distribution of Product Ratings')
plt.xlabel('Stars')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.scatter(df['price/value'], df['stars'], color='orange', alpha=0.5)
plt.title('Relationship between Price and Ratings')
plt.xlabel('Price')
plt.ylabel('Stars')
plt.show()
```

```python
avg_ratings_by_brand = df.groupby('brand')['stars'].mean().sort_values(ascending=False)
plt.bar(avg_ratings_by_brand.index[:10], avg_ratings_by_brand.values[:10], color='green')
plt.title('Average Ratings by Brand (Top 10)')
plt.xlabel('Brand')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns


plt.figure(figsize=(10, 6))
sns.boxplot(x='price/currency', y='price/value', data=df)
plt.title('Price Distribution by Currency')
plt.xlabel('Currency')
plt.ylabel('Price')
plt.show()
```
