```python
# Importing the matplotlib.pyplot module to create plots and visualizations
import matplotlib.pyplot as plt
```

```python
# Creating a DataFrame 'df' from existing data using pandas
df = pd.DataFrame(data)
```

```python
# Create a new figure with a specific size
plt.figure(figsize=(10, 6))

# Plot a histogram of 'IMDB Rating' column from DataFrame 'df'
plt.hist(df['IMDB Rating'], bins=10, color='skyblue', edgecolor='black')

# Set labels for x and y axes
plt.xlabel('IMDB Rating')
plt.ylabel('Frequency')

# Set the title of the plot
plt.title('Distribution of IMDB Ratings')

# Display grid lines on the plot
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Plotting scatter plot for IMDB ratings and movie durations

# Create a new figure with a specific size
plt.figure(figsize=(10, 6))

# Plot a scatter plot of 'IMDB Rating' vs 'Duration' from DataFrame 'df'
plt.scatter(df['Duration'], df['IMDB Rating'], color='skyblue', alpha=0.7)

# Set labels for x and y axes
plt.xlabel('Duration')
plt.ylabel('IMDB Rating')

# Set the title of the plot
plt.title('Scatter Plot of IMDB Ratings vs Movie Durations')

# Display grid lines on the plot
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Calculate the count of movies for each genre
genre_counts = df['Genre'].str.split(', ').explode().value_counts()

# Plotting pie chart for movie genres
plt.figure(figsize=(8, 8))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140)

# Set the title of the pie chart
plt.title('Distribution of Movie Genres')

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Show the plot
plt.show()
```
